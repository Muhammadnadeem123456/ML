```python
# Drop duplicate columns, keeping the first occurrence
dataset = dataset.loc[:, ~dataset.columns.duplicated()]
```

```python
print(dataset.columns.duplicated())
print(dataset.columns[dataset.columns.duplicated()])
```

```python
# Fill NaN values with 0 in the specified columns
fil_na_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[fil_na_zero] = dataset[fil_na_zero].fillna(0)

# Verify the result
print("NaN count after filling:", dataset.isna().sum())
```

```python
fil_na_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']

# Fill NaN values with 0
dataset[fil_na_zero] = dataset[fil_na_zero].fillna(0)

# Verify the result
print(dataset.isna().sum())
```

```python
print(fil_na_zero)
print(dataset[fil_na_zero].head())
```

```python
dataset['Landsize']=dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
print(dataset.isna().sum())
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset=dataset. astype(int)
dataset.head()
```

```python
X = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
re = LinearRegression().fit(X_train, y_train)
```

```python
re.score(X_test, y_test)
```

# Overfiting 

```python
re.score(X_train, y_train)
```

# L1 Lasso Regularization

```python
from sklearn import linear_model
lasso_re = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_re.fit(X_train, y_train)
```

```python
lasso_re.score(X_test, y_test)
```

```python
lasso_re.score(X_train, y_train)
```

# L2 Ridge Regression

```python
from sklearn import linear_model
ridge_re = linear_model.Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_re.fit(X_train, y_train)
```

```python
ridge_re.score(X_train, y_train)
```

```python
ridge_re.score(X_test, y_test)
```
