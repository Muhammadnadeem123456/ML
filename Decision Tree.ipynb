```python
import pandas as pd
df=pd.read_csv('salaries.csv')
df.head()
```

```python
input=df.drop('salary_more_then_100k', axis='columns')
input.head()
```

```python
target=df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
input['company_n']=le_company.fit_transform(input['company'])
input['job_n']=le_job.fit_transform(input['job'])
input['degree_n']=le_degree.fit_transform(input['degree'])
```

```python
input.head()
```

```python
# input.drop(['company', 'job', 'degree'],axis=columns, inplace=True)
# input.head()
input = input.drop(['company', 'job', 'degree'], axis=1, errors='ignore')
input.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(input, target)
```

```python
model.score(input, target)
```

```python
model.predict([[2, 1, 0]])
```

```python
model.predict([[2, 0, 1]])
```

```python
model.predict([[2, 1, 1]])
```

```python
import pandas as pd
```

```python
df1=pd.read_csv('titanic.csv')
df1.head()
```

```python
df1.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df1.head()
```

```python
inputs=df1.drop('Survived', axis='columns')
target=df1.Survived
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
inputs.Age[:20]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.Age[:20]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```
