# Pandas Dummies

```python
df=pd.read_csv('USHome_prices.csv')
df
```

```python
import pandas as pd
dummies=pd.get_dummies(df.town)
dummies=dummies.astype(int)
dummies
```

```python
df
```

```python
df_dummies=pd.concat([df, dummies], axis='columns')
df_dummies
```

# One Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df
```

```python
dfle=df
```

```python
dfle.town = le.fit_transform(dfle.town)
```

```python
dfle
```

```python
X=dfle[['town', 'area']].values
X
```

```python
Y=dfle.price.values
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohes=OneHotEncoder(categorical_features=[0])
```

```python
ohes.fit_transform(X).toarray()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Assuming X is your input data
encoder = ColumnTransformer(
    transformers=[('onehot', OneHotEncoder(sparse_output=False), [0])],  # Apply OneHotEncoder to the first column
    remainder='passthrough'  # Keep the other columns unchanged
)

# Apply the transformation
result = encoder.fit_transform(X)

# Display the result
print(result)
```

```python
# Skip the first column
result_skipped_first_col = result[:, 1:]

# Display the result
print(result_skipped_first_col)
```

```python
encoder.fit(result_skipped_first_col, Y)
```

```python
encoder.transform([[0, 0, 3400]])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Assuming X is your feature matrix and y is the target (prices)
pipeline = Pipeline([
    ('encoder', ColumnTransformer([('onehot', OneHotEncoder(sparse_output=False), [0])], remainder='passthrough')),  # OneHotEncoder for categorical data
    ('model', LinearRegression())  # Regression model
])

# Fit the pipeline
pipeline.fit(X, y)

# Predict price for new data [category_feature, other_feature1, other_feature2, etc.]
# predicted_price = pipeline.predict([[0, 0, 3400]])  # Example new data
# print(predicted_price)
# # Check the number of features in your training data
print(X.shape)  # This will tell you how many features you used during training

# If your dataset `X` has 2 features, change the input for prediction accordingly
predicted_price = pipeline.predict([[0, 3400]])  # Example new data with the correct number of features
print(predicted_price)
```
