```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('nychomeprices.csv')
df
```

# Scatter Plot

```python
%matplotlib inline
plt.xlabel('area(sqr, ft)')
plt.ylabel('prices(US$)')
plt.scatter(df.prices, df.area, marker='+', color='red')
```

# Creating linear regression object so that you can fit data into it fiting means training model using available data points

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']], df.prices)
```

# Now our model is ready to predict the prices

```python
reg.predict(np.array([[5000]]))
```

```python
reg.coef_
```

```python
float(reg.intercept_)
```

# Importing Dataframe with area only predict their prices

```python
import pandas as pd
d=pd.read_csv('areas.csv')
d.head()
```

```python
p=reg.predict(d)
```

```python
d['price']=p
```

# creat another file that contain predicted values in it

```python
d.to_csv('prediction.csv', index=False)
```

# Linear Equation Visual Representation 

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('prices(US$)', fontsize=20)
plt.scatter(df.area, df.prices, marker='+', color='red')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
